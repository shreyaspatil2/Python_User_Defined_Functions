```python
def table_maker(total_attributes):
   
    ''' This function enables you to create a table showing possible combinations of variables based on number of unique values in all columns.
        This created table will be  very usefull for Data Simulation. '''
    
    #Import Libraries
    import numpy as np
    import pandas as pd
    
    #Taking attribute details from user
    Columns = []
    Unique = []
    for i in range(1,total_attributes+1):
        c = input("Enter Name of Attribute " + str(i) +": ")
        Columns.append(c)
        Unique.append(int(input("Enter Number Of Unique Values In "+ c +": ")))
    
    #Create Dataframe
    df = pd.DataFrame()
    
    #Creating first column
    col = [i for i in range(1,Unique[0]+1)]
    for i in Unique[1:]:
        col = col*i
    N = len(col)
    df[Columns[0]] = col
    
    #Creating remaining columns
    for i in range(1,len(Unique)):
        next_col = [j for j in range(1,Unique[i]+1)]
        L = []
        for a in next_col:
            for b in range(np.product(Unique[:i])):
                L.append(a)
        df[Columns[i]] = L*int((N/len(L)))
    
    #Table is ready
    return(df)
    
table_maker(4)
```
